```python
import pandas as pd
import numpy as np

# Load dataset from github
data = pd.read_csv("https://raw.githubusercontent.com/zonination/perceptions/master/probly.csv")
data = (data.melt(id_vars=["text"], 
                  value_vars=["value"], 
                  var_name="key", 
                  value_name="value")
         .assign(text=lambda x: x['text'].str.replace(".", " "))
         .assign(value=lambda x: x['value'].round(0)))

# show data
data.sample(8)
```
