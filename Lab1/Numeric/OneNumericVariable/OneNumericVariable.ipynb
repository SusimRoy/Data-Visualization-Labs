```python
!python -m pip install seaborn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv'
df = pd.read_csv(url)
#  Histogram
sns.histplot(x=df['price'][df['price'] < 300],binwidth=10)
plt.title('Histogram : Bins=10')
plt.show()
sns.histplot(x=df['price'][df['price'] < 300],binwidth=2)
plt.title('Histogram : Bins=2')
plt.show()

## Density 
sns.displot(x=df['price'][df['price'] < 300], kind='kde',bw_adjust=10)
plt.title('Density : Bins=10')
plt.show()
sns.displot(x=df['price'][df['price'] < 300], kind='kde', bw_adjust=2)
plt.title('Density : Bins=2')
plt.show()
```
