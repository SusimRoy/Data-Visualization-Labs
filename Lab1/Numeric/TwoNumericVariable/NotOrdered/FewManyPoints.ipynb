```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Histogram

```python
url = 'https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv'
df = pd.read_csv(url)
df['SalePrice'] = df['SalePrice']/1000

## Histogram
sns.histplot(data=df, x="GrLivArea",binwidth=50)
plt.title('Histogram : GrLivArea')
plt.show()
sns.histplot(data=df, x="SalePrice",binwidth=50)
plt.title('Histogram : SalePrice')
plt.show()
```

# Scatter-Plot

```python
#Raw
sns.scatterplot(data=df, x="GrLivArea", y="SalePrice")
plt.title('Raw')
plt.show()

#Trend
sns.regplot(data=df, x='GrLivArea', y='SalePrice', line_kws={'color': 'g'})
plt.title('Linear Regression')
plt.show()
sns.lmplot(data=df, x='GrLivArea', y='SalePrice', lowess=True)
plt.title('Poly Regression')
plt.show()

#Interactivity
df['SalePrice'] = df['SalePrice']*1000

#Marginal Distribution
sns.jointplot(data=df, x='GrLivArea', y='SalePrice',kind='scatter')
plt.title('Marginal Distribution')
plt.show()

##Dot-Size
df['SalePrice'] = df['SalePrice']/1000
sns.scatterplot(data=df, x="GrLivArea", y="SalePrice",size=0.2)
plt.title('Dot size : 0.2')
plt.show()
```

# Density-Plot

```python
##Desnity-2D : Contour
sns.kdeplot(x=df.GrLivArea, y=df.SalePrice)
plt.title('Density 2D Contour')
plt.show()

##Density-2D : Area
sns.kdeplot(x=df.GrLivArea, y=df.SalePrice, cmap="Reds", fill=True)
plt.title('Density 2D Contour :  Area')
plt.show()

##Hexbin
plt.hexbin(df.GrLivArea, df.SalePrice)
plt.title('Hexbin')
plt.show()

##2D Histogram
plt.hist2d(df.GrLivArea, df.SalePrice, bins=(50,50), cmap=plt.cm.jet)
plt.title('2D Histogram')
plt.show()
```
