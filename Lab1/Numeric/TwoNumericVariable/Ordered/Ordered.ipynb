```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered_comma.csv'
data = pd.read_csv(url)
data
```

# Line-Plot

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='date', y='value', data=data)
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
plt.xticks([2014,2015,2016,2017,2018])
sns.set_style("whitegrid")
plt.grid(True)
plt.show()
```

# Area chart

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="date", y="value", data=data, color="#69b3a2")
plt.fill_between(data.date, data.value, color='#69b3a2', alpha=0.5)
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
plt.xticks([2014,2015,2016,2017,2018])
sns.set_style("whitegrid")
```

# Interactivity

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=data['date'], y=data['value'], fill='tozeroy',fillcolor='#69b3a2',mode='lines',line=dict(color='#69b3a2',width=1)))
fig.update_layout(title="Evolution of Bitcoin price", yaxis_title="bitcoin price ($)")
fig.show()
```

# Connected scatter

```python
data = data.tail(10)
plt.figure(figsize=(10,5))
sns.lineplot(x="date", y="value", data=data, color="#69b3a2")
plt.fill_between(data.date, data.value, color='#69b3a2', alpha=0.5)
plt.scatter(data.date, data.value)
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
sns.set_style("whitegrid")
```

# To cut or not to cut

```python
data = data.tail(10)
plt.figure(figsize=(10,5))
sns.lineplot(x="date", y="value", data=data, color="#69b3a2")
plt.fill_between(data.date, y1=8000, y2=data.value, color='#69b3a2', alpha=0.5)
plt.scatter(data.date, data.value)
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
sns.set_style("whitegrid")
```

# Comparing to a limit

```python
data = data.tail(300)
data["mycolor"] = ["#271569" if x > 7500 else "#69b3a2" for x in data["value"]]
plt.figure(figsize=(10,5))
sns.lineplot(x="date", y="value", data=data, color="black")
plt.fill_between(data.date, y1=7500, y2=data.value, color=data.mycolor)
plt.xticks(ticks = ['2017-07-01','2017-10-01','2018-01-01','2018-04-01'] , labels = ['Jul','Oct','Jan','Apr'])
plt.title("Evolution of Bitcoin price")
plt.ylabel("bitcoin price ($)")
sns.set_style("whitegrid")
plt.legend(labels=["value<7500","value>7500"],loc='center left', bbox_to_anchor=(1, 0.5))
```
